```python
import streamlit as st

import pandas as pd
```

```python
covid_data=pd.read_csv("https://raw.githubusercontent.com/govex/COVID-19/master/data_tables/vaccine_data/global_data/time_series_covid19_vaccine_global.csv")
```

```python
covid_data
```

```python
grouped_data=covid_data.groupby(['Country_Region','Date']).sum()
grouped_data.head()
```

```python
data=grouped_data.reset_index()
data['Date']=data['Date'].astype('datetime64[ns]')
data
```

```python
set(data['Country_Region'])
```

```python


import altair as alt

alt.data_transformers.disable_max_rows()

input_dropdown = alt.binding_select(options=list(data['Country_Region']))

selection = alt.selection_single(fields=['Country_Region'], bind=input_dropdown, name='Country of ')

alt.Chart(data).mark_line().encode(
    x='Date:T',
    y='People_partially_vaccinated:Q',
    color='Country_Region',
    strokeDash='Country_Region',
).add_selection(
    selection
).transform_filter(
    selection
)
```
